```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Load the cleaned dataset (use the name of your uploaded file)
df = pd.read_csv('pizza_sales_clean.csv')

print("✅ Cleaned data loaded successfully!")
df.head()
```

```python
#Step 3 — Check overall structure and missing values
# Inspect data structure and missing values
df.info()
df.isnull().sum()
```

```python
#Step 4 — Revenue by Day of Week
# Total revenue by day of the week
revenue_by_day = df.groupby('order_day')['total_price'].sum().sort_values(ascending=False)

print("Revenue by Day of Week:")
print(revenue_by_day)

# Visualize
revenue_by_day.plot(kind='bar', figsize=(8,4), title='Revenue by Day of the Week', grid=True)
plt.ylabel("Revenue (₹)")
plt.show()
```

```python
#Step 5 — Orders by Hour of Day
orders_by_hour = df.groupby('order_hour')['order_id'].nunique()

print("Orders by Hour of Day:")
print(orders_by_hour)

orders_by_hour.plot(kind='line', marker='o', figsize=(8,4),
                    title='Orders by Hour of the Day', grid=True)
plt.xlabel("Hour")
plt.ylabel("Number of Orders")
plt.show()
```

```python
#Step 6 — Top 10 Best-Selling Pizzas
# By quantity sold
top_pizzas_qty = df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(10)

# By revenue
top_pizzas_rev = df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(10)

print("Top 10 Pizzas by Quantity:")
print(top_pizzas_qty)
print("\nTop 10 Pizzas by Revenue:")
print(top_pizzas_rev)

# Visualize
top_pizzas_qty.plot(kind='barh', figsize=(8,5), title='Top 10 Pizzas by Quantity Sold')
plt.xlabel("Quantity Sold")
plt.show()

top_pizzas_qty = df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(10)

# By revenue
top_pizzas_rev = df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(10)

print("Top 10 Pizzas by Quantity:")
print(top_pizzas_qty)
print("\nTop 10 Pizzas by Revenue:")
print(top_pizzas_rev)

# Visualize
top_pizzas_qty.plot(kind='barh', figsize=(8,5), title='Top 10 Pizzas by Quantity Sold')
plt.xlabel("Quantity Sold")
plt.show()

# By quantity sold
top_pizzas_qty = df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(10)

# By revenue
top_pizzas_rev = df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(10)

print("Top 10 Pizzas by Quantity:")
print(top_pizzas_qty)
print("\nTop 10 Pizzas by Revenue:")
print(top_pizzas_rev)

# Visualize
top_pizzas_qty.plot(kind='barh', figsize=(8,5), title='Top 10 Pizzas by Quantity Sold')
plt.xlabel("Quantity Sold")
plt.show()
```

```python
#Step 7 — Average Order Value (AOV)
aov = df.groupby('order_id')['total_price'].sum().mean()
print(f"💵 Average Order Value (AOV): ₹{aov:.2f}")
```

```python
#Step 8— Estimate Seat Utilization
# Approximate party size (2 pizzas ≈ 1 person)
df['estimated_party'] = np.ceil(df['quantity'] / 2).astype(int)

seats_by_hour = df.groupby('order_hour')['estimated_party'].sum()

seats_by_hour.plot(kind='bar', figsize=(8,4), title='Estimated Seats Occupied by Hour')
plt.ylabel("Approx Seats Used")
plt.show()
```

```python
#Step 9 — Revenue Trend by Date
daily_revenue = df.groupby('order_date')['total_price'].sum()

daily_revenue.plot(figsize=(10,4), title='Daily Revenue Trend', grid=True)
plt.ylabel("Revenue (₹)")
plt.show()
```

```python
#Step 10 — Export Summary Files (For Tableau)
# Save summarized data
revenue_by_day.to_csv('revenue_by_day.csv')
orders_by_hour.to_csv('orders_by_hour.csv')
top_pizzas_qty.to_csv('top_pizzas_by_quantity.csv')
top_pizzas_rev.to_csv('top_pizzas_by_revenue.csv')

print("✅ Summary files saved successfully!")
```
