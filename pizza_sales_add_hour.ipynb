```python
import pandas as pd
from google.colab import files

# Step 1: Upload your CSV file
uploaded = files.upload()

# Step 2: Read the file
df = pd.read_csv('pizza_sales_clean.csv')

# Step 3: Combine order_date and order_time columns correctly
# Your format is like: 2015-01-01 11:57:40 (YYYY-MM-DD HH:MM:SS)
df['order_datetime'] = pd.to_datetime(
    df['order_date'].astype(str) + ' ' + df['order_time'].astype(str),
    format='%Y-%m-%d %H:%M:%S',  # fixed to match your data
    errors='coerce'
)

# Step 4: Extract the hour from the datetime
df['order_hour'] = df['order_datetime'].dt.hour

# Step 5: Verify that it's working
print(df[['order_date', 'order_time', 'order_datetime', 'order_hour']].head(15))

# Step 6: Save and download updated dataset
df.to_csv('pizza_sales_clean_updated.csv', index=False)
files.download('pizza_sales_clean_updated.csv')
```
